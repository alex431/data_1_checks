```python
import pandas as pd
import numpy as np 
```

```python
df=pd.read_csv("assets/vgsales.csv")

df.head()
print(df.dtypes)
#print(df.shape)
#df=df.replace('N/A','null',inplace=True)

df.head()
```

```python
df_2 = df.replace('N/A', None).dropna()
df_2['Platform'] = df_2['Platform'].replace('2600', None)
df_2=df_2.dropna(subset=['Platform']).reset_index(drop=True)

acr_dict={
'Wii':'Nintendo Wii',
'WiiU':'Nintendo Wii U',
'X360':'Microsoft Xbox 360',
'SNES':'Super Nintendo Entertainment System',
'NES':'Nintendo Entertainment System',
'GEN':'Sega Genesis',
'DC':'Sega Dreamcast',
'3DO': '3DO Multiplayer',
'PSV':'Sony Playstation Vita',
'SAT': 'Sega Saturn',
'SCD': 'Sega CD',
'WS': 'Bandai WonderSwan',
'NG': 'SNK Neo Geo',
'TG16': 'NEC TurboGrafx-16t',
'GB': 'Nintendo Game Boy',
'3DS': 'Nintendo 3DS',
'GBA': 'Nintendo Game Boy Advance',
'N64': 'Nintendo 64',
'DS': 'Nintendo DS',
'PS': 'Sony PlayStation',
'PS2': 'Sony PlayStation 2',
'PS3': 'Sony PlayStation 3',
'PS4': 'Sony PlayStation 4',
'XB': 'Microsoft Xbox',
'XOne': 'Microsoft Xbox One',
'GC': 'Nintendo GameCube',
'PSP': 'Sony PlayStation Portable',
'GG': 'Sega Game Gear',
'PCFX': 'NEC PC-FX',
}

df_2['Platform']=df_2['Platform'].replace(acr_dict)
df_2['Year']=df_2['Year'].astype(int)

df_2.head()
df_2.to_csv('Video_Game_sales.csv')

```
